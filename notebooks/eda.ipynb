```python
import pandas as pd
```

```python
train = pd.read_csv('../data/train.csv')
```

```python
train['id'].value_counts().max()
```

```python
# NOTE multilabels
train.drop(['id', 'comment_text'], axis=1).sum(axis=1).value_counts()
```

```python
display(train.drop(['id', 'comment_text'], axis=1).sum(axis=0))
train.drop(['id', 'comment_text'], axis=1).sum(axis=0) / len(train)
```

```python
t = train.drop(['id', 'comment_text'], axis=1).sum(axis=1)
display((t > 0).astype(int).value_counts())
(t > 0).astype(int).value_counts() / len(t)
```
